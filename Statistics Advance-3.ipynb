```python
import numpy as np
from scipy.stats import t

def two_sample_t_test(sample1, sample2):
    """
    Perform two-sample t-test for independent samples.
    
    Parameters:
    sample1 (array-like): First sample data.
    sample2 (array-like): Second sample data.
    
    Returns:
    float: t-statistic
    float: p-value
    """
    # Calculate sample sizes
    n1 = len(sample1)
    n2 = len(sample2)
    
    # Calculate means and variances of the samples
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    
    # Calculate pooled standard deviation
    pooled_std_dev = np.sqrt((var1 + var2) / 2)
    
    # Calculate t-statistic
    t_statistic = (mean1 - mean2) / (pooled_std_dev * np.sqrt(1/n1 + 1/n2))
    
    # Calculate degrees of freedom
    degrees_of_freedom = n1 + n2 - 2
    
    # Calculate p-value
    p_value = 2 * (1 - t.cdf(abs(t_statistic), df=degrees_of_freedom))
    
    return t_statistic, p_value

# Example usage:
sample1 = np.random.normal(loc=50, scale=10, size=30)  # First random sample
sample2 = np.random.normal(loc=55, scale=10, size=30)  # Second random sample

t_statistic, p_value = two_sample_t_test(sample1, sample2)
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```
